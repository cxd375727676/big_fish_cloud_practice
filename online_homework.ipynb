```python
#3.4
from datetime import datetime
dt1=datetime(2018,1,1,2,30,35)
dt2=datetime(2018,2,3,4,20,30)

delta_dt=dt2-dt1
print(dt1)
print(dt2)

print(delta_dt.days)  #相差天数
print(int(delta_dt.total_seconds()/3600))   #相差小时数
```

```python
#4.1
stock1=pd.read_excel("sz50.xlsx",sheet_name="600104.XSHG",index_col=0)
stock2=pd.read_excel("sz50.xlsx",sheet_name="600518.XSHG",index_col=0)
return1=stock1.close.pct_change(5)
return2=stock2.close.pct_change(5)
df=pd.concat([return1,return2],axis=1,keys=["stock1","stock2"])
print(df.tail())
print(df.cov())
```

```python
#4.2
from numpy.random import randint
import matplotlib.pyplot as plt
arr=randint(1,7,(60,6))
count6=np.sum(arr==6,axis=1)
print("平均每轮出现6的次数是%.2f" %(count6.mean()))

#直方图统计
plt.figure(figsize=(10,6))
plt.hist(count6,bins=np.arange(-0.5,7.5))
plt.legend(["BinomialRandomVariable"])
plt.show()
```

```python
#4.3
from talib import abstract
data=pd.read_excel("sz50.xlsx",sheet_name="600030.XSHG",index_col=0)
close=data.close
cci=abstract.CCI(data,timeperiod=10)

fig,axes=plt.subplots(2,1,sharex=True,figsize=(18,8))
axes[0].plot(close)
axes[0].set_title("close")
axes[1].plot(cci)
axes[1].set_title("CCI")
plt.show()
```
